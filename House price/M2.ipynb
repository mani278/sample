# Module 2: Exploration data analysis of visualization and training a model by given attributes

```python
#import library packages
import pandas as p
import matplotlib.pyplot as plt
import seaborn as s
import numpy as n
```

```python
import warnings
warnings.filterwarnings('ignore')
```

#### House price dataset

```python
data = p.read_csv("price.csv")
```

```python
df=data.dropna()
```

```python
df.columns
```

```python
#Histogram Plot 
df['Price'].hist(figsize=(7,6), color='green', alpha=0.7)
plt.xlabel('Price')
plt.ylabel('No of Data')
plt.title('Distribution of Price')
```

```python
#Histogram Plot 
df['Area'].hist(figsize=(7,6), color='green', alpha=0.7)
plt.xlabel('Area')
plt.ylabel('No of Area')
plt.title('Number of Area')
```

```python
#Propagation by variable
def PropByVar(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(10,10), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)
    return n.round(dataframe_pie/df.shape[0]*100,2)
PropByVar(df, 'city')
```

```python
# Heatmap plot diagram
fig, ax = plt.subplots(figsize=(15,10))
s.heatmap(df.corr(), annot=True)
```

```python
df.columns
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(df['Area'],df['Price'])
ax.set_xlabel('Area')
ax.set_ylabel('Price')
plt.show()
```
