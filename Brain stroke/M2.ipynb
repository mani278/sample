```python
# Import the neccessary packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('brain.csv')
```

```python
data.head()
```

```python
df = data.dropna()
```

```python
df['bmi'].hist(figsize= (7,8), color='green')
plt.title("Bmi distribution")
plt.xlabel("Bmi range")
plt.ylabel("Bmi weight")
```

```python
sns.distplot(x = df['age'], bins = 10)
```

```python
sns.countplot(df['target'])
```

```python
sns.boxplot(x = df['age'], y = df['gender'])
```

```python
sns.violinplot(y = df['avg_glucose_level'], x =df['target'])
```

```python
sns.heatmap(df.corr(), annot = True, cmap = 'viridis')
```
